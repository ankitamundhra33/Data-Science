# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,20)
y = x * x
z = x + y
```

```python
plt.subplot(3,2,1)
plt.plot(x,x)
plt.subplot(3,2,2)
plt.plot(x,y)
plt.subplot(3,2,3)
plt.plot(x,z)
plt.subplot(3,2,4)
plt.plot(y,z)
plt.tight_layout()
```

# Subplots using Object Oriented

```python
fig, axes = plt.subplots(2,2)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(1,2)
axes[0] = plt.plot(x,y)
axes[1] = plt.plot(y,x)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(2,2)
axes[(0,1)].plot(x,y)
axes[(1,1)].plot(y,x)
plt.tight_layout()
```

# Setting Labels with Object Oriented Techniques

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x = np.linspace(0,10,5)
y = x * x
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('Graph')
axes.plot(x,y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.4,0.4,0.2,0.3])
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('Large Graph')
axes2.set_xlabel('x-axis')
axes2.set_ylabel('y-axis')
axes2.set_title('Small Graph')
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_xlabel('x-axis')
axes[0].set_ylabel('y-axis')
axes[0].set_title('First')
axes[1].set_xlabel('x-axis')
axes[1].set_ylabel('y-axis')
axes[1].set_title('Second')
plt.tight_layout()
```
